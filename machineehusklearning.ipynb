```python
import pandas as pd
```

```python
heart_disease=pd.read_csv("heart.csv")
heart_disease
```

```python
# create x
x=heart_disease.drop('target',axis=1)
x
# create y 
y=heart_disease['target']
y
```

```python
# chose right machine learning model
from sklearn.ensemble import RandomForestClassifier as clf
clf = clf()  
clf.get_params()
```

```python
# fit model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
clf.fit(x_train,y_train)
```

```python
# evaluate model
predict_y=clf.predict(x_test)
predict_y
```

```python
clf.score(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
# impove model
from sklearn.ensemble import RandomForestClassifier

for i in range(10, 200, 10):
    print(f"Running model with {i} estimators")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train, y_train)  # Use correct n_estimators
    print(f"Accuracy is {clf.score(x_test, y_test)}")
```

```python
# save model 
import pickle
pickle.dump(clf,open("Heart Disease predictor.pkl",'wb'))
```

```python
load_model=pickle.load(open("Heart Disease predictor.pkl",'rb'))
load_model.score(x_test,y_test)
```
